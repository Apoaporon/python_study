```python
import matplotlib.pyplot as plt
# グラフのスタイルを決める
import matplotlib.style
```

```python
# グラフスタイルを決定
matplotlib.style.use("ggplot")
x = [1,2,3]
y = [2,4,9]
plt.plot(x,y)
plt.title('pyplot interface')
plt.show()
```

### オブジェクト指向インターフェース
- 描画オブジェクトに対してサブプロットを追加してサブプロットに対して描画を行う

```python
x = [1,2,3]
y = [2,4,9]
# 描画オブジェクト fig とサブプロット axを生成
fig, ax = plt.subplots()

ax.plot(x,y)
ax.set_title('object-oriented interface')
plt.show()
```

```python
# 2つのsubplotを生成
fig, axes = plt.subplots(2)
plt.show()
```

```python
# 2 * 2のsubplotを生成
fig, axes = plt.subplots(2, 2)
plt.show()
```

```python
# 1行 * 2列のsubplotを生成
fig, axes = plt.subplots(ncols=2)
plt.show()
```

```python
fig, axes = plt.subplots(ncols=2)
# サブプロットにタイトルを設定
axes[0].set_title('subplot title 0')
axes[1].set_title('subplot title 1')
# 描画オブジェクトにタイトルを設定
fig.suptitle("figure title")
plt.show()
```

```python
fig, ax = plt.subplots()
# 軸のラベルをつける
ax.set_xlabel('x label')
ax.set_ylabel('y label')
# 凡例用のラベルをつける
ax.plot([1,2,3], [2,4,9], label='legend label')
# 凡例の位置を指定する
ax.legend(loc='best') # その他：lower right 右下　upper left, center left ...
plt.show()
```

```python
# ファイルの出力
fig, ax = plt.subplots()
ax.set_title("subplot title")
fig.savefig("sample-figure.png") # png形式で取得
fig.savefig("sample-figure.svg") # svg形式で保存
```
