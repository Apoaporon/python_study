```python
# ヒストグラムを生成
rng = np.random.default_rng(123)
mu = 100 # 平均値
sigma = 15 # 標準偏差
# 平均と標準偏差を指定し、その範囲で指定個数要素を作成
x = rng.normal(mu, sigma, 1000)
fig, ax = plt.subplots()
# n:各ビン（棒）の要素数、
# bins: ビンの境界の値
n,bins,patches = ax.hist(x)
print(f"n is {n}")
print(f"bins is {bins}")
plt.show()
```

```python
# 円グラフ
labels = ['spam', 'ham', 'egg']
x = [10, 3, 1]
fig,ax = plt.subplots()
# デフォルトは反時計回り, 3時の位置が初期値、  startangle=90, counterclock=Falseで時計回りに
ax.pie(x, labels=labels, startangle=90, counterclock=False) # pie:円グラフ
plt.show()
```

```python
# 円グラフに影をつける
fig,ax = plt.subplots()
# shadowで影をつける、autopctで割合を％で表示
ax.pie(x, labels=labels, startangle=90, counterclock=False, shadow=True, autopct='%1.2f%%')
plt.show()
```

```python
# 円グラフを一部切り抜いて強調する方法
explode = [0, 0.2, 0] # 1番目の要素を切り出す
fig, ax = plt.subplots()

ax.pie(x, labels=labels, startangle=90, counterclock=False, shadow=True, autopct='%1.2f%%', explode=explode)
plt.show()
```

```python

labels = ['A', 'B', 'C']
x = [6, 3, 1]
explode = [0, 0.2, 0]
fig, ax = plt.subplots()
# pieメソッドの引数「autopct='%1.1f%%'」で、「60.0%」などの割合を表示しています。「%1.1f%%」は左から順番に以下を指定しています。
# % → 書式化演算子
# 1 → 最小桁数
# .1f → 小数点以下の桁数
# % → 書式化演算子
# % → 「%」の表示
ax.pie(x, 
       labels=labels, 
       startangle=90, # 3時の位置から90度回転
       counterclock=False, # 本来は反時計回りだが、Falseなので時計回り
       autopct='%1.1f%%', 
       explode=explode)
ax.axis('equal')
```

```python
# ヒストグラムに折れ線グラフを追加して近侍曲線を描画
rng = np.random.default_rng(123)
x = rng.normal(size=1000)
fig, ax = plt.subplots()
# ヒストグラム
counts, edges, patches = ax.hist(x, bins=25)
# 近侍曲線に用いる点を求めるヒストグラムのビンの中点
print(counts)
print(edges)
print(edges[:-1]) # 末尾の要素以外
print(edges[1:]) # 最初の要素以外
# 各ビンの中点を取得
x_fit = (edges[:-1]+edges[1:]) / 2
print(f"x_fit is {x_fit}")
# 近侍曲線をプロット
y = 1000 * np.diff(edges) * np.exp(-x_fit ** 2 / 2) / np.sqrt(2 * np.pi)
ax.plot(x_fit, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)
mu = 100
sigma = 15
x = np.random.normal(mu, sigma, 1000)
fig, ax = plt.subplots()
n, bins, patches = ax.hist(x, bins=25, orientation='horizontal')
print(len(n), len(bins))
for i, num in enumerate(n):
    print('{:.2f} - {:.2f} {}'.format(bins[i], bins[i + 1], num))

plt.show()
```

```python
# 線のスタイルを指定
fig, ax = plt.subplots()
# linewidth : 線の太さを指定
# linestyle : 線の種類を指定
ax.plot([1,3], [3,1], linestyle='--',linewidth=5.5, label='--, 5.5')
ax.plot([1,3], [1,3], linestyle='-.',linewidth=10, label='-. , 10')
ax.plot([1,3], [2,2], linestyle=':',linewidth=3, label='. , 3')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('xlabel', family='fantasy', size=20)
ax.set_ylabel('ylabel', family='serif', size=30, weight='light')
ax.set_title('graph title', famliy='monospace', size=25, weight='heavy')
plt.show()
```

```python
# フォントのスタイルを辞書で定義
fontdict={
    'family':'fantasy',
    'size':20,
    'weight':'normal',
}
fig,ax=plt.subplots()
# テキストも描画可能
ax.text(0.2, 0.4, 'Text', size=20)
# 辞書で指定したフォントを指定
ax.set_xlabel('xlabel', fontdict=fontdict)
ax.set_ylabel('ylabel', fontdict=fontdict)
# 個別指定したsizeで上書き
ax.set_title('graph title', fontdict=fontdict, size=40)
plt.show()
```
