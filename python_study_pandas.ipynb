#### データフレームデータの抽出
- loc
    - インデックス名, カラム名を指定する
- iloc
    - index番号, カラム番号を指定する

```python
# 全抽出
display(df.loc[:, :])
# 列だけを指定
display(df.loc[:, "A"])
# 列を複数指定
display(df.loc[:, ["A", "B"]])
# 行名を指定
display(df.loc["01", :])
# 行名を複数指定
display(df.loc[["01","03"],:])
# 行名、列名両方指定
display(df.loc[["01","03"],:"A"])
```

```python
# 2行目2列目の要素を抽出
display(df.iloc[1,1])
# 2行目以降、2列目の要素を抽出
display(df.iloc[1:, 1])
# 2行目以降, 2列目までの要素を抽出
display(df.iloc[1:, :2])
```
